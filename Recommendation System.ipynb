# Recommendation Systems

Recommender systems, also known as recommendation systems or engines, are type of software applications designed to provide personalized suggestions or recommendations to users. These systems use algorithms and machine learning techniques to analyze large datasets, learn user preferences, and generate personalized items recommendations.

Such systems aim to enhance user experience by reducing search time, increasing engagement, providing personlized suggestions, and increasing user satisfactions.

### Various applications

- Product recommendations (Amazon, eBay)
- Movie & TV show recommendations (Netflix)
- Music recommendation (Spotify)
- Social network recommendations (Facebook, Instagram)
- News & article recommendations (Google News)

### User behavior tracking

Recommender systems track various aspects of user behavior, preferences, and interactions. The specific aspects that are tracked can vary depending on the type of the recommender system and the platform it is implemented on.

Here are some commonly tracked aspects of user behavior and interactions:
##### 1. Explicit Feedback
- Ratings and reviews: Users providing explicit ratings or reviews for items(movies, products, books, etc.).
##### 2. Implicit Feedback
- Click and view history: Tracking which items users viewed or interacted with.
- Purchase history: Analyzing items that users have bought in the past.
- Search queries: Understanding what users are searching for within the platform.
##### 3. User Demographics
- Age, gender, location: To better understand user characteristics.
- User profiles: Aggregated information about user's perferences, interests, and behavior over time.
##### 4. Contextual Information
- Time of day: Recommending different items based on the time of day.
- Device type: Recommending items based on the user's current location.
- Location: Adapting recommendations for different devices.
##### 5. Social Interactions
- Social network connections and activity: Analyzing user's connections and content preferences on social media platforms.
##### 6. User Engagement Metrics
- Frequency of visits: How often user interacts with the platform.
- Session duration: How long a user spends on the platform during a session.
##### 7. Feedback on Recommendations
- Did user accept or ignore previous reccomendations
- Post-engagement feedback: Collecting feedback on wheter the user found the recommendation helpful or not.
##### 8. Browsing History
- Pages visited: Tracking the user's navigation through the platform.
- Items added to cart: Identfying items that users have expressed interest in but not necessarily purchased.

### Types of Recommender Systems

There are three main types of Recommender Systems

### Collaborative Filtering
Collaborative filtering methods make predictions about a user's preferences by collecting preferences and behavior of a group of users (collaborating). It recommends items based on the similarity between users or items. If user A likes items X, Y, and Z, and user B likes items X and Y, then the system will recommend item Z to user B.

The underlying idea is that users who have similar tastes or preferences in the past are likely to have similar tastes in the future.

<img src="collaborative_filtering.png" width="70%" style="float: left;" />

Collaborative filtering can be broadly categorized into two main types: user-based and item-based collaborative filtering.

#### User-Based Collaborative Filtering

This approach relies on the similarity between users. If two users have liked or interacted with similar items in the past, it is assumed that they share common preferences.

#### Item-Based Filtering

This approach focuses on the similarity between items. If two items have been liked or interacted with by similar sets of users, they are considered similar.

<img src="user_vs_item_based.png" width="90%" style="float: left;" />

### Content-Based Filtering
Content-based filtering is another common approach used in recommender systems, focusing on the characteristics or features of items and users to make recommendations. Unlike collaborative filtering, content-based filtering does not rely on user interactions with items or the similarities betwee users. Instead, it recommends items based on the content and attributes of the items themselves and similiraty between them.

If item A has features X,Y, and Z, and item B has features X and Y, then the system will recommend item A to a user who likes features X and Y.

Base assumption is that users like items with similar attributes or features.

<img src="content-based_filtering.png" width="70%" style="float: left;" />

### Hybrid Filtering
Hybrid Filtering is Recommender System that combines Collaborative Filtering and Content-Based Filtering to overcome the limitations of each approach and improve recommendation accuracy.

### Popularity based recommendation

#### Weight rating

A simple and commonly used formula for weighting ratings based on popularity is to use a combination of the average rating and the number of reviews. This approach accounts for both the quality (average rating) and the popularity (number of reviews) of a product.*

$$
\text{Weighted Rating} = \left( \frac{{\text{Number of Reviews}}}{{\text{Number of Reviews} + \text{Minimum Reviews Threshold}}} \right) \times \text{Average Rating} + \left( \frac{{\text{Minimum Reviews Threshold}}}{{\text{Number of Reviews} + \text{Min Reviews Threshold}}} \right) \times \text{Global Average Rating}
$$

- **Number of Reviews:** The actual number of reviews for a specific product.
- **Average Rating:** The average rating of the product.
- **Minimum Reviews Threshold:** A threshold value to avoid products with very few reviews having an overly dominant effect.
- **Global Average Rating:** The average rating across all products in the dataset.

```python
review_counts = ratings['ratings_count'].astype('int')
```

```python
# Threshold for minimum reviews required for product to be listed - top 5% products by ratings count
quantile_threshold = 0.95

min_reviews_threshold = review_counts.quantile(quantile_threshold)

print("Minimum review count required based on the 95th percentile: "+ str(min_reviews_threshold))
```

```python
# Global average ranking
global_average = ratings['ratings_average'].mean()

print ("Average rating across all products in the dataset: " + str(global_average))
```

```python
# Weighted rating

def weighted_rating(df):
    numberOfReviews = df['ratings_count']
    minimumReviewsThreshold = min_reviews_threshold
    averageRating = df['ratings_average']
    globalAverageRating = global_average

    weighted_rating = (numberOfReviews / (numberOfReviews + minimumReviewsThreshold) * averageRating) + (minimumReviewsThreshold / (minimumReviewsThreshold + numberOfReviews) * globalAverageRating)

    return weighted_rating
```

```python
qualified_reviews = ratings[(ratings['ratings_count'] >= 28)]
qualified_reviews.sort_values(by='ratings_count', ascending=False)
```

```python
qualified_reviews['weighted_rating'] = qualified_reviews.apply(weighted_rating, axis=1)
```

```python
qualified_reviews = qualified_reviews.sort_values ('weighted_rating', ascending=False)
```
